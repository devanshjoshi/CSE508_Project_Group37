```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("cnbc_headlines.csv")
data
```

```python
df = data.copy()
df["Tweet"] = df["Headlines"] + " " + df["Description"]
df.dropna(inplace = True)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer


new_words = {
    'crushes': 10,
    'beats': 5,
    'misses': -5,
    'trouble': -10,
    'falls': -100,
}
vader = SentimentIntensityAnalyzer()
vader.lexicon.update(new_words)
```

```python
df["Filtered_Date"] = pd.to_datetime(df.Time).dt.date
sentiment_scores = df['Tweet'].apply(vader.polarity_scores)
# df["Filtered_Date"] = pd.to_datetime(df.Time).dt.date
df
```

```python
type(sentiment_scores)
```

```python
sentiment_scores
```

```python
df["neg"] = 0.0
df["neu"] = 0.0
df["pos"] = 0.0
df["compound"] = 0.0
df["index"] = 0
```

```python
df.index[0]
```

```python
index_arr = df.index
count = 0
for i in index_arr:
    row = sentiment_scores[i]
    neg, neu, pos, compound = row["neg"], row["neu"], row["pos"], row["compound"]
    df.at[i, "index"] = count
    count += 1
    df.at[i, "neg"] = neg
    df.at[i, "pos"] = pos
    df.at[i, "neu"] = neu
    df.at[i, "compound"] = compound
df
    
```
